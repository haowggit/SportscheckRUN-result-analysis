```python
import pandas as pd
```

```python
df = pd.read_csv("race_results.csv", index_col=0, sep=";") 
df['AK'] = df['AK'].str.replace('.', '', regex=False)
df[['AK_Integer', 'AK_Rest']] = df['AK'].str.extract(r'(\d+)(.*)')
df.rename(columns={'AK_Integer': 'Place in Group'}, inplace=True)
df.rename(columns={'AK_Rest': 'Group name'}, inplace=True)
df.drop(columns=['AK'], inplace=True)
df.index = df.index.astype(int)
df['Jg.'] = df['Jg.'].fillna(0).astype(int)
df['Group name'] = df['Group name'].replace('', 'divers')
df.to_csv("processed_race_results.csv", sep=";", index=True)
```

```python

grouped_counts = df.groupby('Group name').size()
grouped_counts.to_csv("grouped_counts.csv", sep=";")
```
